```python
# from Chinese to English 

import pandas as pd
import numpy as np

reference = pd.read_csv("NTCIR-13_MedWeb_en_training.csv")
candidate_from_zh = pd.read_csv("NTCIR-13_MedWeb_en_from_zh_amazon_training.csv")

# remove punctuation inside
import re
def  clean_text_en(df, text_field): 
    df[text_field] = df[text_field].str.lower()
    df[text_field] = df[text_field].apply(lambda elem: re.sub(r"(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)|^rt|http.+?", "", elem))  
    return df

# update

reference = clean_text_en(reference,"Tweet")
candidate_from_zh = clean_text_en(candidate_from_zh,"Tweet")

print('Processing text dataset >>>')
from nltk.tokenize import RegexpTokenizer

tokenizer = RegexpTokenizer(r'\w+')
def text_to_wordlist(text, lower=False):
    # Tokenize
    text = tokenizer.tokenize(text)
    return text

def process_comments(list_sentences, lower=False):
    comments = []
    for text in list_sentences:
        txt = text_to_wordlist(text, lower=lower)
        comments.append(txt)
    return comments

reference = list(reference.Tweet)
candidate_from_zh = list(candidate_from_zh.Tweet)

# update

reference = process_comments(reference)
candidate_from_zh = process_comments(candidate_from_zh)

# evaluate BLEU

from nltk.translate.bleu_score import sentence_bleu

score = []
total_num = len(reference)

for i in range (0,total_num):
    metric = sentence_bleu(reference[i],candidate_from_zh[i],weights=(1,0))
    score.append(metric)

# correct sentences in new file

import csv

src_file_name1 = "NTCIR-13_MedWeb_en_training.csv"
tran_file_name2 = "NTCIR-13_MedWeb_en_from_zh_amazon_training.csv"  # 源文件路径
gen_file_name = "update_bleu0.1_en_from_zh_training.csv"  # 目标文件路径

total_num = len(score)

with open(src_file_name1, 'r', encoding='UTF-8') as f1, open(tran_file_name2, 'r', encoding='UTF-8') as f2, open(gen_file_name, 'w', newline='') as f:
    rows = csv.reader(f1)
    rows_tran = csv.reader(f2)
    write = csv.writer(f)
    write.writerow(["ID","Tweet","Influenza","Diarrhea","Hayfever","Cough","Headache","Fever","Runnynose","Cold"])
    for i in range(total_num):
        if score[i] <= 0.1:
            for index, row in enumerate(rows):
                if index == i+1:
                    write.writerow(row)
                    f1.seek(0)
                    break
        else:
            for index, row in enumerate(rows_tran):
                if index == i+1:
                    write.writerow(row)
                    f2.seek(0)
                    break
```

```python
# from English to Japanese

reference = pd.read_csv("NTCIR-13_MedWeb_ja_training.csv", na_values='NaN',keep_default_na=False)
candidate_from_en = pd.read_csv("NTCIR-13_MedWeb_ja_from_en_amazon_training.csv")

# remove punctuation inside

import re
from zhon.hanzi import punctuation
print('Processing text dataset >>>')

def clean_text_ja(df, text_field):
    df[text_field] = df[text_field].str.lower()
    df[text_field] = df[text_field].apply(lambda elem:re.sub(r'[{}]+'.format(punctuation),'',elem))
    return df

# update 
reference = clean_text_ja(reference,"Tweet")
candidate_from_en = clean_text_ja(candidate_from_en,"Tweet")

# setup Japanese tokenizer
import nagisa

def tokenize_jp(doc):
    doc = nagisa.tagging(doc)
    return doc.words

# update
reference = [list(tokenize_jp(i)) for i in reference.Tweet]
candidate_from_en = [list(tokenize_jp(i)) for i in candidate_from_en.Tweet]

# evaluate BLEU

from nltk.translate.bleu_score import sentence_bleu

score = []
total_num = len(reference)

for i in range (0,total_num):
    metric = sentence_bleu(reference[i],candidate_from_en[i],weights=(1,0))
    score.append(metric)
    
# correct sentences in new file

import csv

src_file_name1 = "NTCIR-13_MedWeb_ja_training.csv"
tran_file_name2 = "NTCIR-13_MedWeb_ja_from_en_amazon_training.csv"  # 源文件路径
gen_file_name = "update_bleu0.65_ja_from_en_training.csv"  # 目标文件路径

total_num = len(score)

with open(src_file_name1, 'r', encoding='UTF-8') as f1, open(tran_file_name2, 'r', encoding='UTF-8') as f2, open(gen_file_name, 'w', newline='') as f:
    rows = csv.reader(f1)
    rows_tran = csv.reader(f2)
    write = csv.writer(f)
    write.writerow(["ID","Tweet","Influenza","Diarrhea","Hayfever","Cough","Headache","Fever","Runnynose","Cold"])
    for i in range(total_num):
        if score[i] <= 0.65:
            for index, row in enumerate(rows):
                if index == i+1:
                    write.writerow(row)
                    f1.seek(0)
                    break
        else:
            for index, row in enumerate(rows_tran):
                if index == i+1:
                    write.writerow(row)
                    f2.seek(0)
                    break
```
